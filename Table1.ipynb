```python
import pandas as pd

#print(pd.__version__)

# Create a dataframe
df = pd.DataFrame(
    {
        "Fruit": ["Apples", "Bananas","Grapefruits","Strawberries"],
        'Weight':[1,3,500,250],
        "Unit":['kg','kg','g','g','mg'],
        'Price per kg':[42,36,50,99],
        "Currency":['CZK', "CZK", 'CZK','CZK']
    }
)
df

```

```python
# Modify the price for apples to 45 CZK and weight to 2kg
df.loc[df['Fruit']=='Apples','Price per kg']=45
df.loc[df['Fruit']=='Apples','Weight']=2
df
```

```python
#Create a new column Total with calculation weight x price
df["Total"]= df['Weight']*df['Price per kg']
df
```

```python
#Create a new column Incorrect column
df['Incorrect column']=pd.NaT
df
```

```python
#Delete the new column
del df['Incorrect column']
df
```

```python
#Create a new column Mssage
df['Mssage']=pd.NaT
df
```

```python
#Update the name to Message
df=df.rename(columns={'Mssage':'Message'})
df
```

```python
#Put the text inside the Message column
df['Message'] = "In your basket you have " + df['Weight'].map(str) + df['Unit'].map(str) + ' of ' + df['Fruit'] + ' for ' + df['Price per kg'].map(str) + df['Currency'].map(str)

df['Message2'] = df.agg('In your basket you have {0[Weight]} {0[Unit]} of {0[Fruit]} for {0[Price per kg]} {0[Currency]}'.format, axis=1)
df
```
